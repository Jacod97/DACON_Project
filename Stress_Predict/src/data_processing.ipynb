```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import os
```

```python
path = "../data/"
train_path = os.path.join(path, "train.csv")
test_path = os.path.join(path, "test.csv")

train = pd.read_csv(train_path, encoding='utf-8')
test = pd.read_csv(test_path, encoding='utf-8')
display(train.head())
print(list(train.columns))
```

```python
train['BMI'] = round(train['weight']*10000 / (train['height'] ** 2),1)
train['BMI']
```

```python
train.describe()
```

```python
def bmi_category(bmi):
    """성인 기준 WHO BMI 판정 함수"""
    if bmi < 18.5:
        return "underweight"
    elif 18.5 <= bmi < 25:
        return "normal"
    elif 25 <= bmi < 30:
        return "overweight"
    elif 30 <= bmi < 35:
        return "obesity class I"
    elif 35 <= bmi < 40:
        return "obesity class II"
    else:
        return "obesity class III"

train['BMI_category'] = train["BMI"].apply(lambda x: bmi_category(x))
train
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 차원축소에 사용할 컬럼 선택
pca_features = [
    "cholesterol",
    "systolic_blood_pressure",
    "diastolic_blood_pressure",
    "glucose",
    "bone_density"
]

# 스케일링 (PCA는 스케일 차이에 민감함)
scaler = StandardScaler()
scaled_train = scaler.fit_transform(train[pca_features])
scaled_test = scaler.transform(test[pca_features])

# PCA: 2차원으로 축소
pca = PCA(n_components=3, random_state=42)
train_pca = pca.fit_transform(scaled_train)
test_pca = pca.transform(scaled_test)

# 결과를 데이터프레임에 합치기
train["health_pca1"] = train_pca[:, 0]
train["health_pca2"] = train_pca[:, 1]
train["health_pca3"] = train_pca[:, 2]
test["health_pca1"] = test_pca[:, 0]
test["health_pca2"] = test_pca[:, 1]
test["health_pca3"] = test_pca[:, 2]

print("PCA 설명 분산 비율:", pca.explained_variance_ratio_)
```
